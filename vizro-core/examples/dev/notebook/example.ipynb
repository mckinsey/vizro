```python
import vizro.models as vm
from vizro import Vizro
```

```python
card = vm.Card(
    id="card_1",
    text="Card 1",
)

card2 = vm.Card(
    id="card_2",
    text="Card 2",
)

page = vm.Page(
    # id="page_1",
    title="Page 1",
    components=[card, card2],
)

dashboard = vm.Dashboard(
    id="dashboard_1",
    title="Dashboard 1",
    pages=[page],
)
```

```python
from dash._callback_context import context_value

context_value.set({})

dashboard = vm.Dashboard.model_validate(dashboard, context={"build_tree": True})
app = Vizro().build(dashboard)

# TODO: Make this work!
app.run(debug=False)
```
