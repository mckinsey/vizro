```python
import vizro.models as vm
from vizro import Vizro
```

```python
card = vm.Card(
    id="card_1",
    text="Card 1",
)

page = vm.Page(
    # id="page_1",
    title="Page 1",
    components=[card],
)

dashboard = vm.Dashboard(
    id="dashboard_1",
    title="Dashboard 1",
    pages=[page],
)
```
